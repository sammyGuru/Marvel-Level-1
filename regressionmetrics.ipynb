## Regression Metrics:


*   Regression is used for continuous target values. Metrics used in these cases should be able to work on continuous values.
The following are the different metrics used in regression:

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df= pd.read_csv("/content/Housing.csv")
df.head()
df = pd.get_dummies(df)#one - hot encoding: converting categorical data variables from strings to numeric data
```

```python
X = df.drop('price', axis = 1)
Y = df['price']
X_train , X_test, Y_train, Y_test = train_test_split(X , Y , train_size = 0.8, random_state = 42)

model = LinearRegression()
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

1. Mean Absolute Error
* It is calculated by taking the average of the absolute differnces between the predicted and the actual values
* It is sensitive to outliers.

```python
mae = mean_absolute_error(Y_test, y_pred)
print("Mean absolute error =", mae )
```



2.   MSE(Mean Squared Error):
* To find the average squared error between the predicted and actual values.
RMSE can also be used instead of MSE.
* Squaring the errors penalises large erorrs.
1_vkgzQYMaYeM71bmBYuufUw.webp


```python
mse = mean_squared_error(Y_test,y_pred)
print("mean squared error =", mse)
```

3. Root Mean Squared Error:
* RMSE provides a measure of the spread of the errors around the mean prediction.
* It is commonly used when you want to penalize large errors but still maintain the same scale as the original data.
insights_rmse_formula.jpg

```python
rmse = mean_squared_error(Y_test, y_pred, squared = False)
print("root mean squared error = ", rmse)
```

4. R2 Squared Method:
* R-squared (R2) is a statistical measure that represents the proportion of the variance in the dependent variable that is explained by the independent variables in a regression model
* A high R2 value suggests that the model captures most of the underlying patterns in the data, while a low R2 value may indicate that the model is missing important explanatory factors or suffers from underfitting.
1_D9wPS2odLYsNOvOLffOMCw.png

```python
r2 = r2_score(Y_test, y_pred)
print(r2)#0.65 means that 65% of the variability in the dependent variable is explained by the independent variables.
```
