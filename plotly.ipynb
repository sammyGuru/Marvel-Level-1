```python
#heat maps
a = np.arange(0, 100, 3)
b = np.arange(0, 100, 4)

[X , Y]=np.meshgrid(a,b)
Z = np.cos(X/2)+ np.sin(Y/4)

fig = go.Figure(data = go.Heatmap(x = a,y=b,z=Z ))
fig.show()
```

```python
df = px.data.tips()

# plotting the figure
fig = px.line_3d(df, x="sex", y="day",
                 z="time", color="sex")

fig.show()
```

```python
df = px.data.iris()

fig = px.scatter_3d(df, x = 'sepal_width',
                    y = 'sepal_length',
                    z = 'petal_width',
                    color = 'species')

fig.show()
```

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

fig.show()
```
