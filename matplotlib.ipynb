```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y = 4 + 2*np.sin(2*x)
fig, ax = plt.subplots()
ax.plot(x, y, linewidth = 2.0)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),#xticks and yticks set the data points on the axis
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.xlabel('x')
plt.ylabel('2*sin(x)')
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig , ax = plt.subplots()
#plotting and setting labels
ax.plot(x,y1,label = 'sin(x)')
ax.plot(x,y2,label = 'cos(x)')
#setting limits
ax.set_xlim(0,10)
ax.set_ylim(-1,1)

ax.set_xticks(np.arange(0, 11, 2))
ax.set_yticks(np.arange(-1, 1.1, 0.5))
ax.legend()

plt.savefig('plots.png')

plt.show()
```

```python
iris = pd.read_csv('/content/iris.csv')
```

```python
print(iris)
```

```python
# Define features and target variable
features = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
target = 'variety'

# Create subplots for each feature
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axes = axes.flatten()# flatten the 2D array of subplot axes into a 1D array.

for i, feature in enumerate(features):
    ax = axes[i]
    ax.scatter(iris[feature], iris[target])
    ax.set_xlabel(feature)
    ax.set_ylabel(target)
    ax.set_title(f'Scatter Plot of {feature} vs {target}')

plt.show()
```

```python
# Define features and target variable
features = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
target = 'variety'

# Create subplots for each feature
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axes = axes.flatten()# flatten the 2D array of subplot axes into a 1D array.

for i, feature in enumerate(features):
    ax = axes[i]
    sizes = iris[feature] * 50 #multiply the value of each feature by 50 to scale it appropriately and adjust the size of the bubbles.
    ax.scatter(iris[feature], iris[target], s=sizes,alpha = 0.5)#alpha-adjusts transparency, s represents the size which ddepicts the value of each feature
    ax.set_xlabel(feature)
    ax.set_ylabel(target)
    ax.set_title(f'Scatter Plot of {feature} vs {target}')

plt.show()
```

```python
#bar plot
#1. simple bar plot
x = ['mon','tue','wed','thur','fri']
y1 = [3,5,6,7,4]

plt.subplot(1,3,1)
plt.bar(x,y1)
plt.xlabel('day of the week')
plt.ylabel('no of hours')


#2. grouped
x = np.arange(len(x))
y2 = [2,5,7,3,6]
width = 0.33
plt.subplot(1,3,2)

plt.bar(x - width/2, y1, width, label="group1", color = 'red' )
plt.bar(x + width/2, y2, width, label="group2", color = 'blue' )


#3. stacked
plt.subplot(1,3,3)
plt.bar(x,y1,label = 'group1')
plt.bar(x,y2,bottom = y1, label = 'group2')

plt.show()
```

```python
#1.histogram- to show the distribution of a continuous variable
x = np.random.randn(1000)
plt.subplot(3,1,1)
plt.hist(x, bins = 20, color='black' )#bins means that the histogram is divided into 30 equal intervals
#2. box plot - show the distribution of a dataset, specifically its median, quartiles, and outliers.
plt.subplot(3,1,2)
plt.boxplot(x)
plt.title('Box Plot')

#violin plot
plt.subplot(3,1,3)
plt.violinplot(x)
plt.tight_layout()
plt.show()
```

```python
# 3. pie plt
x = ['mon','tue','wed','thur','fri']
y = [3,5,6,7,4]
plt.pie(y,labels = x,autopct ='%1.1f%%', startangle=140 )#autopct-pecifies the format of the percentage labels displayed on each slice.
#tartangle=140 specifies the angle at which the first slice starts.
```

```python
#contour plots
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))#generates 256 equally spaced points between -3 and 3 along both the x-axis and y-axis.
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)#function combines polynomial terms with exponential decay to produce a surface with complex shapes and contours.
levels = np.linspace(Z.min(), Z.max(), 7)#evenly spaced 7 levels

fig, ax = plt.subplots()

ax.contourf(X, Y, Z, levels=levels)#contour() for no fill

plt.show()
```

```python
#marginal plots
np.random.seed(0)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
plt.subplot(2,2,3)
plt.scatter(x,y,alpha = 0.5)

#Create marginal histograms on top and right side
plt.subplot(2,2,1)
plt.hist(x, bins=30, color='blue', alpha=0.5, density=True)
plt.gca().get_yaxis().set_visible(False)#hide y axis

plt.subplot(2,2,4)
plt.hist(x, bins=30, color='red', alpha=0.5,orientation='horizontal', density=True)
plt.gca().get_xaxis().set_visible(False)#hide x axis
plt.tight_layout()
plt.show()
```
